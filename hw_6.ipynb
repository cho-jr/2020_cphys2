```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_sample_image
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
china = load_sample_image("china.jpg")
```

```python
image = china/225
plt.imshow(image)
```

```python
X = image.reshape(-1, 3)
kmeans = KMeans(n_clusters=2).fit(X)
```

```python
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
```

```python
plt.imshow(segmented_img)
```

```python
kiner = {}
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    kiner[k] = kmeans.inertia_

print(kmeans.inertia_)
plt.plot(list(kiner.keys()), list(kiner.values()))
```

n_clusters 10 이상은 의미 없음
기울기 크게 변하는 지점3, 6 


```python
from pprint import pprint
kiner_value = list(kiner.values())
pprint(kiner_value)
len(kiner_value)
```

```python
d_value = []

for k in range(1, len(kiner_value)-1):
    d_add = kiner_value[k] - kiner_value[k-1]
    d_value.append(d_add)
    
plt.plot(range(1, len(kiner_value)-1), d_value)
    
    
```
